```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras import optimizers
from tensorflow.keras.layers import Input, Dense, Flatten

import time
```

# 흑백 이미지 crop 

```python
(train_x, train_y), (test_x, test_y) = tf.keras.datasets.mnist.load_data()
```

```python
plt.imshow(train_x[0], cmap=plt.cm.binary)
print(train_y[0])
```

```python
image = train_x[0]
print(image.shape)
```

```python
cropped = image[5:20, 5:20]
plt.imshow(cropped, cmap="binary")
plt.show()
```

# 파일로 저장/읽기

```python
plt.imsave("cropped_mnist.jpg", cropped)
```

```python
loaded = plt.imread("cropped_mnist.jpg")
print(loaded.shape)
plt.imshow(loaded)
plt.show()
```

# 컬러 영상

```python
(train_x, train_y), (test_x, test_y) = tf.keras.datasets.cifar10.load_data()
```

```python
image = train_x[77]
plt.imshow(image)
plt.show()
```

```python
print(image.shape)
```

```python
cropped = image[9:25, 1:30, :]
plt.imshow(cropped)
plt.show()
```

# 파일로 저장/읽기

```python
plt.imsave("cropped.jpg", cropped)
```

```python
loaded = plt.imread("cropped.jpg")
print(loaded.shape)
plt.imshow(loaded)
plt.show()
```

# numpy로 저장/읽기

```python
np.save("crooped.npy", cropped)
```

```python
loaded = np.load("crooped.npy")
print(loaded.shape)
plt.imshow(loaded, cmap="binary")
plt.show()
```
